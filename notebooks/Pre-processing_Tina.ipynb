# Pre-process

```python
train_df = pd.read_csv('../data/interim/train.csv')
test_df = pd.read_csv('../data/interim/test.csv')
val_df = pd.read_csv('../data/interim/validate.csv')
```

```python
train_df
```

```python

target_col = ['beer_style']
all_features_list = [col for col in train_df.columns if col not in target_col]
features_cate_col = ['brewery_name']
features_numeric_col = [col for col in all_features_list if col not in features_cate_col]

X_train_df = train_df.drop(columns=target_col)
y_train_df = train_df[target_col]
```

```python
print (f'before drop NA and target shape {train_df.shape}.')
print (f'after drop NA and target shape {X_train_df.shape}.')
print(f'Diff is {train_df.shape[0]-X_train_df.shape[0]} rows.')
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
scaler = StandardScaler()
encoder = OneHotEncoder()
```

```python
# save the scaler for later
from joblib import dump
dump(scaler, '../models/scaler.joblib')
dump(encoder, '../models/encoder.joblib')
```

```python
from joblib import load
import pandas as pd

numeric_transformer = load('../models/scaler.joblib')
categorical_transformer = load('../models/encoder.joblib')
```

```python
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, features_cate_col),
         ('num', numeric_transformer, features_numeric_col)
     ])
```

```python
cat_feature_names = [] 
X_preprocessed = preprocessor.fit_transform(X_train_df)
```

```python
print("Shape of X_preprocessed:", X_preprocessed.shape)
```

```python
numeric_feature_names = numeric_transformer.get_feature_names_out(input_features=features_numeric_col)
num_numeric_features = len(numeric_feature_names) # 5

cat_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(input_features=features_cate_col)
num_cat_features = len(cat_feature_names) # 4901

column_names = list(cat_feature_names) + list(numeric_feature_names)
```

```python
print(f'First 5 columns names are: {column_names[:5]}.')
print(f'Last 5 columns names are: {column_names[-5:]}.')
```

```python
print("Length of column_names:", len(column_names)) # 4906 by the X_preprocessed has 4906 columns
print("Length of numeric_feature_names:", len(numeric_feature_names)) # 5
print("Length of cat_feature_names:", len(cat_feature_names)) # 4901
print("Shape of X_preprocessed:", X_preprocessed.shape) #(1063169, 4906)
```

```python
# convert to a df
X_dense = X_preprocessed.toarray()
X_preprocessed_df = pd.DataFrame(X_dense, columns=column_names)
```
